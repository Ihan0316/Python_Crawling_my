```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import urllib.request


query = '맛조개'
driver = webdriver.Chrome()
driver.get(f'https://www.google.com/imghp')
# 구글 이미지 검색창 이름이 : q
search_bar = driver.find_element(By.NAME,"q")
# 검색 키워드를 춘식이 설정
search_bar.send_keys(query)
# 전달.
search_bar.submit()
# 잠시 대기 시간 ms 1000ms = 1초
PAUSE_TIME = 2
# 현재 문서의 스크롤의 높이를 표시.
last_hegiht = driver.execute_script("return window.scrollY")
print(f"처음 last_hegiht : {last_hegiht}")
new_height = 0

interval = 2
prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
    #화면 가장 아래로 스크롤 내리기
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(interval)

    curr_height = driver.execute_script("return document.body.scrollHeight")
    if curr_height == prev_height:
        break

    prev_height = curr_height

img_elements = driver.find_elements(
    By.TAG_NAME,"img")
# 다시 이미지만 추출해서 담음 임시 리스트
print(f"img_elements ================================ : {img_elements}")

save_path = f'D:\crawl-501\{query}'
import os
if not os.path.exists(save_path):
    os.mkdir(save_path)


save_path = f'D:\crawl-501\{query}'
imgs = []
# enumerate 함수 이용하면 해당 인덱스도 같이 추출 가능.
for idx, img in enumerate(img_elements):
    print(f"{query} - {idx+1}/{len(img_elements)} 진행 중...")

    try:
        img_src = img.get_attribute("src")
        img_alt = img.get_attribute("alt")

        if img_src and "http" in img_src:  # src가 비어 있거나 base64가 아닌 경우 저장
            imgs.append({"alt": img_alt, "src": img_src})
            urllib.request.urlretrieve(img_src, f"{save_path}/{query}_{idx}.png")
            print(f"✅ {idx}: 다운로드 완료 - {img_alt}")

    except Exception as e:
        print(f"❌ {idx}: 다운로드 실패 - {e}")
        pass
driver.close()


print('done')
```
